```python
import pandas as pd
import json
```

```python
with open("../experiments/results_transe_wn11.json", "r") as f:
    transe = json.load(f)

with open("../experiments/results_distmult_wn11.json", "r") as f:
    distmult = json.load(f)

with open("../experiments/results_complex_wn11.json", "r") as f:
    complex = json.load(f)
    
df = pd.concat((
    pd.DataFrame([{'model_name': i['model_name'], **i['model_params'], **i['results']} for i in transe]).assign(model="transe"),
    pd.DataFrame([{'model_name': i['model_name'], **i['model_params'], **i['results']} for i in distmult]).assign(model="distmult"),
    pd.DataFrame([{'model_name': i['model_name'], **i['model_params'], **i['results']} for i in complex]).assign(model="complex"),
))

df.sort_values("mrr", ascending=False)
```
